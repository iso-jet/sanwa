```python
import pandas as pd
import glob
import codecs
from pandas import Series, DataFrame
```

```python
glob.glob('*.csv')
```

```python
data1 = pd.read_csv("Stress.csv",encoding='shift_jis')
data1
```

```python
data1 = data1[data1["Category"] == "Sustain"]
```

```python
for col_name in data1.columns:
    if col_name != "Point" and col_name != "Category" and col_name != "Stress" and col_name != "Ratio":
        data1 = data1.drop(col_name,axis=1)
```

```python
data1
```

```python
data2 = pd.read_csv("Support.csv",encoding='shift_jis')
```

```python
for col_name in data2.columns:
    if col_name != "Point" and col_name != "Type" :
        data2 = data2.drop(col_name,axis=1)
        
data2
```

```python
data2 = data2.drop_duplicates()
data2
```

```python
data_stress_type = pd.merge(data1,data2,how = "outer",left_on = 'Point',right_on='Point')
data_stress_type
```

```python
data_series = Series(data_stress_type['Type'])
pd.Series.value_counts(data_series)
```

```python
data_stress_type.at[data_stress_type['Type'] == "Guide",'Type'] = 1
data_stress_type.at[data_stress_type['Type'] == "V - Stop",'Type'] = 2
data_test = data_stress_type.fillna(0)
data_series = Series(data_test['Type'])
pd.Series.value_counts(data_series)
```

```python
data_test.corr()
```

```python
#データの分類
from sklearn.model_selection import train_test_split
#決定木
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

#説明変数と目的変数
X = data_test.drop(['Point','Category','Type'],axis=1)
Y = data_test['Type']
```

```python
#データを分ける
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state = 50)

#決定木インスタンス（エントロピー、深さ5）
tree_model = DecisionTreeClassifier(criterion='entropy',max_depth=5,random_state=50)
```

```python
#fitting
tree_model.fit(X_train,Y_train)

print("train:",tree_model.__class__.__name__,tree_model.score(X_train,Y_train))
print("test:",tree_model.__class__.__name__ , tree_model.score(X_test,Y_test))
```
