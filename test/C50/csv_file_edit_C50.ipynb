```python
#SVM
from sklearn.svm import LinearSVC

#標準化のためのモジュール
from sklearn.preprocessing import StandardScaler

#説明変数と目的変数
X = data_test_press.drop(['From','Point','Type'],axis=1)
Y = data_test_press['Type']

#データを分ける
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state = 50)
```

```python
#model
model = LinearSVC()

#標準化
sc = StandardScaler()
sc.fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

clf = model.fit(X_train_std,Y_train)

print("train:",clf.__class__.__name__ ,clf.score(X_train_std,Y_train))
print("test:",clf.__class__.__name__ , clf.score(X_test_std,Y_test))
```
